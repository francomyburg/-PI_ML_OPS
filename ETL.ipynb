<h3>importando pandas y cargando los datasets</h3>

```python
import pandas as pd
```

```python
dfa=pd.read_csv("data/amazon_prime_titles.csv")
```

```python
dfa.head(3)
```

```python
dfd=pd.read_csv("data/disney_plus_titles.csv")
dfh=pd.read_csv("data/hulu_titles.csv")
dfn=pd.read_csv("data/netflix_titles.csv")
```

- Creando columna id

```python
dfa["id"]="a"+dfa["show_id"]
dfd["id"]="d"+dfd["show_id"]
dfh["id"]="h"+dfh["show_id"]
dfn["id"]="n"+dfn["show_id"]
```

- rellenar nulos de la columna rating con G

```python
dfa.rating.fillna("G",inplace=True)
dfd.rating.fillna("G",inplace=True)
dfh.rating.fillna("G",inplace=True)
dfn.rating.fillna("G",inplace=True)
```

- Cambiar formato de fechas

```python
dfa.date_added=pd.to_datetime(dfa.date_added,infer_datetime_format=True,)
dfd.date_added=pd.to_datetime(dfd.date_added,infer_datetime_format=True)
dfh.date_added=pd.to_datetime(dfh.date_added,infer_datetime_format=True)
dfn.date_added=pd.to_datetime(dfn.date_added,infer_datetime_format=True)
```

- crear columnas duration_int y duration type

```python
dfa[["duration_int","duration_type"]]=dfa["duration"].str.split(" ",expand=True)
dfd[["duration_int","duration_type"]]=dfd["duration"].str.split(" ",expand=True)
dfh[["duration_int","duration_type"]]=dfh["duration"].str.split(" ",expand=True)
dfn[["duration_int","duration_type"]]=dfn["duration"].str.split(" ",expand=True)
dfa.drop(columns="duration",inplace=True)
dfd.drop(columns="duration",inplace=True)
dfh.drop(columns="duration",inplace=True)
dfn.drop(columns="duration",inplace=True)
```

```python
dfn.dtypes
```

- Corrigiendo los tipos de datos

```python
#COLUMNA duration_int en formato de dato tipo int
dfa["duration_int"]=dfa["duration_int"].astype("Int64")
dfd["duration_int"]=dfd["duration_int"].astype("Int64")
dfh["duration_int"]=dfh["duration_int"].astype("Int64")
dfn["duration_int"]=dfn["duration_int"].astype("Int64")
```

- Textos en minusculas

```python
#amazon
dfa["title"]=dfa["title"].str.lower()
dfa["type"]=dfa["type"].str.lower()
dfa["director"]=dfa["director"].str.lower()
dfa["cast"]=dfa["cast"].str.lower()
dfa["country"]=dfa["country"].str.lower()
dfa["rating"]=dfa["rating"].str.lower()
dfa["listed_in"]=dfa["listed_in"].str.lower()
dfa["description"]=dfa["description"].str.lower()
dfa["duration_type"]=dfa["duration_type"].str.lower()
#disney
dfd["title"]=dfd["title"].str.lower()
dfd["type"]=dfd["type"].str.lower()
dfd["director"]=dfd["director"].str.lower()
dfd["cast"]=dfd["cast"].str.lower()
dfd["country"]=dfd["country"].str.lower()
dfd["rating"]=dfd["rating"].str.lower()
dfd["listed_in"]=dfd["listed_in"].str.lower()
dfd["description"]=dfd["description"].str.lower()
dfd["duration_type"]=dfd["duration_type"].str.lower()
#hulu
dfh["title"]=dfh["title"].str.lower()
dfh["type"]=dfh["type"].str.lower()
dfh["director"]=dfh["director"].str.lower()
#dfh["cast"]=dfh["cast"].str.lower()
dfh["country"]=dfh["country"].str.lower()
dfh["rating"]=dfh["rating"].str.lower()
dfh["listed_in"]=dfh["listed_in"].str.lower()
dfh["description"]=dfh["description"].str.lower()
dfh["duration_type"]=dfh["duration_type"].str.lower()
#netflix
dfn["title"]=dfn["title"].str.lower()
dfn["type"]=dfn["type"].str.lower()
dfn["director"]=dfn["director"].str.lower()
dfn["cast"]=dfn["cast"].str.lower()
dfn["country"]=dfn["country"].str.lower()
dfn["rating"]=dfn["rating"].str.lower()
dfn["listed_in"]=dfn["listed_in"].str.lower()
dfn["description"]=dfn["description"].str.lower()
dfn["duration_type"]=dfn["duration_type"].str.lower()
```

```python
dfn.head()
```

- Unir dataframes

```python
datamov=pd.concat([dfa,dfd,dfh,dfn])
```

- dataframes de ratings

```python
r1=pd.read_csv("data/ratings/1.csv")
r2=pd.read_csv("data/ratings/2.csv")
r3=pd.read_csv("data/ratings/3.csv")
r4=pd.read_csv("data/ratings/4.csv")
r5=pd.read_csv("data/ratings/5.csv")
r6=pd.read_csv("data/ratings/6.csv")
r7=pd.read_csv("data/ratings/7.csv")
r8=pd.read_csv("data/ratings/8.csv")
```

```python
r1.head()
```

- Cambiar el formato de fecha y unir

```python
r1['timestamp']=pd.to_datetime(r1['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r1.timestamp.astype("datetime64[ns]")
r2['timestamp']=pd.to_datetime(r2['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r2.timestamp.astype("datetime64[ns]")
r3['timestamp']=pd.to_datetime(r3['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r3.timestamp.astype("datetime64[ns]")
r4['timestamp']=pd.to_datetime(r4['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r4.timestamp.astype("datetime64[ns]")
r5['timestamp']=pd.to_datetime(r5['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r5.timestamp.astype("datetime64[ns]")
r6['timestamp']=pd.to_datetime(r6['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r6.timestamp.astype("datetime64[ns]")
r7['timestamp']=pd.to_datetime(r7['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r7.timestamp.astype("datetime64[ns]")
r8['timestamp']=pd.to_datetime(r8['timestamp'],unit='s').dt.strftime('%Y-%m-%d')
r8.timestamp.astype("datetime64[ns]")
```

```python
rating=pd.concat([r1,r2,r3,r4,r5,r6,r7,r8])
```

```python
#cambiar nombre de ratings por score
rating.rename(columns={"rating":"score"},inplace=True)
```

```python
rating.duplicated().value_counts()
```

```python
rating.drop_duplicates(inplace=True)
```

- Obtener el promedio de los scores por cada movieId

```python
df=rating.groupby(["movieId"]).score.mean().round(decimals=1)
```

```python
#formato dataframe
df=df.to_frame()
```

```python
df.reset_index(inplace=True)
```

```python
df.score.value_counts()
```

```python
df.rename(columns={"movieId":"id"},inplace=True)
```

<p align="center">unir dataframe de ratings con movies</p>

```python
datafinal=pd.merge(datamov,df,on="id")
datafinal.rename(columns={"cast":"casting"},inplace=True)
datafinal.to_csv(path_or_buf="datfinal.csv")
```
